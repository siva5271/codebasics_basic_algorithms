```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.preprocessing import OneHotEncoder
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('/home/siva/Downloads/py-master/ML/16_regularization/Melbourne_housing_FULL.csv')
df
```

```python
df=df.dropna(subset=['Price'],axis=0)
df.reset_index(inplace=True,drop=True)
df.info()
```

```python
df['Bedroom2'].fillna(df['Bedroom2'].median(),inplace=True)
df['Bathroom'].fillna(df['Bathroom'].median(),inplace=True)
df['Car'].fillna(0,inplace=True)
df['Landsize'].replace(0.0,np.NaN,inplace=True)
df['Landsize'].fillna(df['Landsize'].mean(),inplace=True)
df.drop(['BuildingArea','YearBuilt','Lattitude','Lattitude',\
    'Address','SellerG','Date','Distance','Postcode','Car'],axis=1,inplace=True)
df
```

```python
df.drop(['Longtitude','Method','Propertycount'],axis=1,\
        inplace=True)
df
```

```python
ohe=OneHotEncoder()
suburb=ohe.fit_transform(df[['Suburb']])
suburb_temp=pd.DataFrame(suburb.toarray(),columns=ohe.categories_)
suburb_temp.drop(['Abbotsford'],inplace=True,axis=1)
suburb_temp
```

```python
ohe=OneHotEncoder()
Type=ohe.fit_transform(df[['Type']])
type_temp=pd.DataFrame(Type.toarray(),columns=ohe.categories_)
type_temp.drop(['h'],inplace=True,axis=1)
type_temp
```

```python
ohe=OneHotEncoder()
council_area=ohe.fit_transform(df[['CouncilArea']])
council_area_temp=pd.DataFrame(council_area.toarray(),columns=ohe.categories_)
council_area_temp.drop(['Banyule City Council'],axis=1,inplace=True)
council_area_temp
```

```python
ohe=OneHotEncoder()
region_name=ohe.fit_transform(df[['Regionname']])
region_name_temp=pd.DataFrame(region_name.toarray(),columns=ohe.categories_)
region_name_temp.drop(['Eastern Metropolitan'],axis=1,inplace=True)
region_name_temp
```

```python
df=pd.concat([df,suburb_temp,type_temp,council_area_temp,region_name_temp],axis=1)
df
```

```python
# df.drop(['Suburb','Type','CouncilArea','Regionname'],\
#         axis=1,inplace=True)
df=df.reset_index(drop=True)
df
```

```python
X=df.drop(['Price'],axis=1)
y=df['Price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train
```

```python
model=Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
model.fit(X_train,y_train)
```
