```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('/home/siva/Downloads/heart_failure_clinical_records_dataset.csv')
df
```

```python
df['age_z']=(df['age']-df['age'].mean())/df['age'].std()
df
```

```python
df=df[(df['age_z']<3) | (df['age_z']<-3)]
df
```

```python
z_scores=np.abs(stats.zscore(df))
z_scores
```

```python
df=df[(z_scores<3).all(axis=1)]
df.reset_index(inplace=True,drop=True)
df
```

```python
scaler=MinMaxScaler()
df_scaled=scaler.fit_transform(df)
df_scaled=pd.DataFrame(df_scaled,columns=df.columns)
df_scaled
```

```python
models={
    'svc':{
        'model':SVC(gamma='auto'),
        'params':{
            'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
            'C':[1,10,20]
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[50,100,250,500,1000,1500]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(),
        'params':{
            'C':[1,10,20]
        }
    },
    'gaussian_nb':{
        'model':GaussianNB(),
        'params':{
            
        }
    },
    'decision_tree':{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion':['gini', 'entropy', 'log_loss']
        }
    }
}
```

```python
score=[]
for model_name,model_params in models.items():
    gsv_clv=GridSearchCV(model_params['model'],\
        model_params['params'],cv=5,return_train_score=False)
    gsv_clv.fit(df_scaled.drop(['DEATH_EVENT'],axis=1),\
                df_scaled['DEATH_EVENT'])
    score.append([model_name,gsv_clv.best_score_,\
                  gsv_clv.best_params_])
score
```

```python
df_score=pd.DataFrame(score,columns=['model','best_score','best_params'])
df_score
```

```python
pca=PCA(n_components=0.95)
```

```python
df_pcaed=pca.fit_transform(df_scaled.drop(['DEATH_EVENT'],axis=1))
df_pcaed
```

```python
score_pcaed=[]
for model_name,model_params in models.items():
    gsv_clv=GridSearchCV(model_params['model'],\
        model_params['params'],cv=5,return_train_score=False)
    gsv_clv.fit(df_pcaed,\
                df_scaled['DEATH_EVENT'])
    score_pcaed.append([model_name,gsv_clv.best_score_,\
                  gsv_clv.best_params_])
score_pcaed
```

```python
df_pcaed_score=pd.DataFrame(score_pcaed,columns=['model','score','param'])
df_pcaed_score
```
